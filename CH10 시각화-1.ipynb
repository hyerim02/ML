## <span style="color:lightpink">pyplot 모듈의 이해</span> 

- 직선 그리기

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
```

## <span style="color:lightpink">Figure, Axes 이해</span> 

Figure: 그림판 크기 조절
Axes: 실제 그림 그리는 method를 가짐 (범례, 타이틀, X축, Y축 등등)

## <span style="color:lightpink">Figure, Axes 활용</span> 

```python
plt.figure(figsize=(10,4))
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
```

```python
plt.figure(figsize=(4,4),facecolor='skyblue')
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(4,4))
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
```

## <span style="color:lightpink">plot으로 선 그래프 그리기</span> 

```python
import numpy as np
import pandas as pd
```

```python
x_value=[1,2,3,4]
y_value=[2,4,6,8]
plt.plot(x_value,y_value)
plt.show()
```

```python
plt.plot(x_value,y_value,color='blue')
```

```python
plt.plot(x_value,y_value,color='red',marker='o',linestyle='dashed',linewidth=2,markersize=10)
```

## <span style="color:lightpink">축 명칭 설정, 축의 눈금값 회전, 범례 설정하기</span> 

```python
plt.plot(x_value,y_value,color='red',marker='o',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x_value=np.arange(0,100)
y_value=2*x_value
plt.plot(x_value,y_value,color='blue')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(rotation=90)
plt.show()
```

```python
x_value=np.arange(0,100)
y_value=2*x_value
plt.plot(x_value,y_value,color='blue')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(ticks=np.arange(0,105,5),rotation=90)
plt.show()
```

```python
x_value=np.arange(0,100)
y_value_1=2*x_value
y_value_2=4*x_value
plt.plot(x_value,y_value_1,color='blue',label='blue')
plt.plot(x_value,y_value_2,color='orange',label='orange')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.xticks(ticks=np.arange(0,105,5),rotation=90)
plt.show()
```

```python
x_value=np.arange(0,10)
y_value_1=2*x_value
plt.plot(x_value,y_value_1,color='red',marker='o',linestyle='dashed',linewidth=2,label='red')
plt.bar(x_value,y_value_1,color='blue',label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
figure = plt.figure() 
ax = plt.axes()
ax.plot(x_value, y_value_1, color='red',  marker='o', linestyle='dashed', label='red')
ax.bar(x_value, y_value_1, color='blue', label='bar plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.legend()
plt.show()
```

## <span style="color:lightpink">여러개의 subplots을 가지는 Figure를 생성하고 여기에 개별 그래프를 시각화</span> 

- nrows가 1일 때는 튜플로 axes를 받을 수 있음.  
- nrows나 ncols가 1일때는 1차원 배열형태로, nrows와 ncols가 1보다 클때는 2차원 배열형태로 axes를 추출

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))
```

```python
print('ax type:', type(ax))
print('ax[0] type:', type(ax[0]))
print('ax[1] type:', type(ax[1]))
```

```python
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')

ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

ax[0].legend()
ax[1].legend()

ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend() 
ax[1][0].legend()
ax[1][1].legend() 

plt.show()
```
